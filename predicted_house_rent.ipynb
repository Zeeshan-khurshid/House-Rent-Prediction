```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("houseprice.csv")
df
```

```python
reg= linear_model.LinearRegression()
```

```python
reg.fit(df[['area']],df.price)
```

```python
%matplotlib inline
plt.scatter(df.area,df.price,color='blue',marker='D')
plt.xlabel("area")
plt.ylabel("price")
plt.title('Price of House')
plt.plot(df.area,reg.predict(df[['area']]),marker='*',color='black')
```

```python
reg.predict([[3300]])
```
